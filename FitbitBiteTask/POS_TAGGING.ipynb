# Parts of Speech tags using Spacy Library

```python
#Perform Standard Imports
import spacy
import nltk
nlp = spacy.load('en_core_web_sm')#loading english language library
import pandas as pd
```

# 1. open csv file

```python
row = []
with open('bbc-news.txt',encoding = "utf-8") as f:
    doc = nlp(f.read())
    #doc = pd.DataFrame(row,columns = ['Text','','value'])
```

# To find the POS

```python
for token in doc:
    print(f"{token.text:{10}}{token.pos_:{10}}")
```

```python
#data = pd.DataFrame(row,columns = ['POS','Text','Count'])
#data.set_index("POS", inplace = True) 
#data
```

```python
from collections import Counter

nouns = []
adjectives = []
verb = []


for token in doc:
    if token.pos_ == 'NOUN':
        nouns.append(token)
        token.pos_ = 'NOUN'
        c1 = Counter(([token.pos for token in doc if token.pos_ == 'NOUN']))
        df1 = {'Text':nouns}, {'Count':c1}
        df = pd.DataFrame(data = df1)
    if token.pos_ == 'ADJ':
        adjectives.append(token)
        c2 = Counter(([token.pos_ for token in doc if token.pos_ == 'ADJ']))
        df2 = {'Text':adjectives}, {'Count':c2}
        df4 = pd.DataFrame(data = df2)
#df2 = df1.append({'Text':adjectives}, {'Count':c2})
    if(token.pos_ == 'VERB'):
        verb.append(token)
        c3 = Counter(([token.pos_ for token in doc if token.pos_ == 'VERB']))
        df3 = {'Text':verb}, {'Count':c3}
        df5 = pd.DataFrame(data = df3)
horizontal_stack = pd.concat([df,df4,df5],axis = 0)
```
